```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data_sales.csv",delimiter=';')
data.head()
data.info()
data['tanggal_penjualan'] = pd.to_datetime(data['tanggal_penjualan'])
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
describe = data.describe()
print (describe)
```

```python
# Assuming 'data' is a pandas DataFrame with 'total_harga' as one of its columns
plt.subplot(2, 2, 2)
plt.hist(data['total Harga'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Total Pembelian')
plt.xlabel('Total Pembelian')
plt.ylabel('Frequency')

plt.show()
```

```python
# Bar Plot penjualan by Jenis Barang
grouped_data = data.groupby('jenis_barang')['jumlah'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(grouped_data['jenis_barang'], grouped_data['jumlah'], color='skyblue', edgecolor='black')
plt.title('Total Penjualan by Jenis Barang')
plt.xlabel('Jenis Barang')
plt.ylabel('Total Harga')
```

```python
# Bar Plot Pendapatan by Jenis Barang
grouped_data = data.groupby('jenis_barang')['total Harga'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(grouped_data['jenis_barang'], grouped_data['total Harga'], color='skyblue', edgecolor='black')
plt.title('Total Pendapatan by Jenis Barang')
plt.xlabel('Jenis Barang')
plt.ylabel('Total Harga')
```

```python
# Barplot Distribusi Jenis Kelamin
gender_data = ['pria', 'wanita']
count_data = [len(data) // 2] * 2  # Membagi data menjadi dua kelompok

# Membuat barplot
plt.subplot(2, 2, 4)
plt.bar(gender_data, count_data, color=['blue', 'pink'])
plt.title('Barplot Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah')
```

```python
#Bar Plot Total Penjualan by Jenis Kelamin
grouped_data = data.groupby('jenis_kelamin_pelanggan')['jumlah'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(grouped_data['jenis_kelamin_pelanggan'], grouped_data['jumlah'], color=['blue', 'pink'], edgecolor='black')
plt.title('Total Penjualan by Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Total Harga')
```

```python
#Bar Plot Total Penjualan by Jenis Kelamin
grouped_data = data.groupby('jenis_kelamin_pelanggan')['total Harga'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(grouped_data['jenis_kelamin_pelanggan'], grouped_data['total Harga'], color=['blue', 'pink'], edgecolor='black')
plt.title('Total Pendapatan by Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Total Harga')
```

```python
# Scatter Plot
plt.subplot(2, 2, 1)
plt.scatter(data['jumlah'],data['total Harga'], color='blue', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('Jumlah Barang')
plt.ylabel('Total Harga')
```

```python
# Scatter Plot
plt.subplot(2, 2, 1)
plt.scatter(data['jenis_barang'],data['total Harga'], color='blue', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('Jenis Barang')
plt.ylabel('Total Harga')
```
